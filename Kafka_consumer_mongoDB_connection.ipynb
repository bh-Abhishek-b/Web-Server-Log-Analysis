```python
import json
from pymongo import  MongoClient
from kafka import KafkaConsumer
```

# Connecting to the MongoDB Database

```python
try:
   client = MongoClient('localhost',27017)
   db = client.AnomalyDetection
   print("Connected successfully!")
except:  
   print("Could not connect to MongoDB")
```

```python
db.list_collection_names()
```

```python
def json_deserializer(data):
    return json.loads(data)
```

```python
consumer = KafkaConsumer('topic3',bootstrap_servers=['localhost:9092'],auto_offset_reset="latest",group_id="group-1",
                       value_deserializer=json_deserializer,consumer_timeout_ms=15000)
```

# Taking the results from Consumer and inserting the data to the MongoDB Database

```python
for msg in consumer:
    record = json.loads(msg.value) 
    HostIP = record['Host']
    Date = record['Date']
    Time = record['Time']
    Method = record['Method']
    Endpoint = record['Endpoint']
    Protocol= record['Protocol']
    Status_Code= record['Status Code']
    Content_Size= record['Content Size']
    No_of_Requests= record['No of Requests']
    Anomaly= record['Anomaly']
    Anomaly_Score= record['Anomaly_Score']
    
    if record['Anomaly']==1:
    # Create dictionary and ingest data into MongoDB
        try:
           rec = {'Host':HostIP,'Date':Date,'Time':Time,'Method':Method,'Endpoint' :Endpoint,'Protocol': Protocol,'Status_Code':Status_Code,'Content_Size':Content_Size,'No_of_Requests':No_of_Requests,'Anomaly':Anomaly,'Anomaly_Score':Anomaly_Score }
           rec_id1 = db.Details.insert_one(rec)
           print("Data inserted with record ids", rec_id1)
        except:
           print("Could not insert into MongoDB")
```
