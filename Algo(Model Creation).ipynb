```python
import pandas as pd
```

```python
df=pd.read_csv('third_result.csv')
```

```python
df
```

```python
anomalydetected=df[df.Anomaly==1]
anomaltnotdetected=df[df.Anomaly==0]
```

```python
print(anomalydetected.shape)
print(anomaltnotdetected.shape)
```

```python
anotheranomalynotdetected_downsampling=df.loc[df.Anomaly==0][:287]
```

As the data is imbalanced,we need to downsample the data which contains majority of values

# Downsampling the dataset

Using resample method

```python
from sklearn.utils import resample
```

```python
anomaltnotdetected_downsample=resample(anomaltnotdetected,replace=True,n_samples=len(anomalydetected),random_state=94)
```

```python
anomaltnotdetected_downsample.shape
```

```python
downsampled_df=pd.concat([anomaltnotdetected_downsample,anomalydetected])
```

```python
downsampled_df=downsampled_df.sample(frac=1,random_state=94)
```

Using manual method

```python
downsampled_df2=pd.concat([anotheranomalynotdetected_downsampling,anomalydetected])
```

```python
downsampled_df2=downsampled_df2.sample(frac=1,random_state=94)
```

```python
downsampled_df
```

```python
downsampled_df2
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from xgboost import XGBRFClassifier,XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier,AdaBoostClassifier
```

```python
X=downsampled_df2.drop(['Date','Time','Anomaly_Score','Anomaly'],axis=1)
```

```python
Y=downsampled_df2.Anomaly
```

```python
Y
```

# Spliting the Data for Training and Testing

```python
(x_train,x_test,y_train,y_test)=train_test_split(X,Y,test_size=0.3,random_state=94)
```

```python
x_train
```

# Parameters of optimization of Models

```python
param1={"random_state":[i for i in range (100)],"n_estimators":[i for i in range(200)],"n_jobs":[i for i in range(20)],"learning_rate":[0.1,0.2,0.5,0.001,0.01,0.02]}
param2={'max_depth': [3,4,5],'n_estimators':[100, 200, 300],"learning_rate": [0.02,0.1,0.01,0.2,0.001,0.002]}
param3={"random_state":[i for i in range (100)],"n_estimators":[i for i in range(200)],"criterion":["squared_error","friedman_mse"],"learning_rate":[0.001,0.002,0.2,0.1,0.01]}
param4={"random_state":[i for i in range (100)],"n_estimators":[i for i in range(200)],"criterion":["entropy","gini"],"n_jobs":[i for i in range(1,20)]}
param5={"random_state":[i for i in range (100)],"n_estimators":[i for i in range(200)],"learning_rate":[0.001,0.002,0.2,0.1,0.01]}
```

# Using Randomized Search of Hyperparameter Optimization on every model

```python
model1=RandomizedSearchCV(XGBClassifier(),param_distributions=param1,cv=5,verbose=1,scoring="accuracy")
model2=RandomizedSearchCV(XGBRFClassifier(),param_distributions=param1,cv=5,verbose=1,scoring='accuracy')
model3=RandomizedSearchCV(CatBoostClassifier(),param_distributions=param2,cv=5,verbose=1,scoring='accuracy')
model4=RandomizedSearchCV(GradientBoostingClassifier(),param_distributions=param3,verbose=1,cv=5,scoring='accuracy')
model5=RandomizedSearchCV(RandomForestClassifier(),param_distributions=param4,verbose=1,cv=5,scoring='accuracy')
model6=RandomizedSearchCV(AdaBoostClassifier(),param_distributions=param5,verbose=1,cv=5,scoring='accuracy')
```

# Fitting the data on each Model

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
model4.fit(x_train,y_train)
model5.fit(x_train,y_train)
model6.fit(x_train,y_train)
```

# Each Models Best Parameters found

```python
print(model1.best_params_)
print(model2.best_params_)
print(model3.best_params_)
print(model4.best_params_)
print(model5.best_params_)
print(model6.best_params_)
```

# Predicting the Anomalies

```python
p=model1.predict(x_test)
p2=model2.predict(x_test)
p3=model3.predict(x_test)
p4=model4.predict(x_test)
p5=model5.predict(x_test)
p6=model6.predict(x_test)
```
